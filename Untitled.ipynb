```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load Dataset
try:
    df = pd.read_csv("sales.csv")
    print("Dataset losded successfully!\n")
except FileNotFoundError:
    print("Error: Filenot found. Pleasecheck the filename.")
except Exception as e:
    print(f"Error: {e}")

#Show first few rows
print("First 5 rows:")
print(df.head(), "\n")

# Explore structure     
print("Dataset Info:")
print(df.info(), "\n")

print("Missing Values:")
print(df.isnull().sum(), "\n")

#Clean dataset (drop missing values if any)
df = df.dropna()

# Summary statistics
print("Summary Statistics:")
print(df.describe(), "\n")

# Grouping: Average revenue per region
region_group = df.groupby("Region")["Revenue"].mean()
print("Average Revenue ny Region:")
print(region_group, "\n")

# Grouping: Average revenue per product
product_group = df.groupby("Product")["Revenue"].mean()
print("Average Revenue by Product:")
print(product_group, "\n")

plt.figure(figsize=(8,5))
plt.plot(df["Date"], df["Revenue"], marker="0", linestyle="-", color="b", label="Revenue")
plt.title("Sales Revenue Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue ($)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

region_group.plot(kind="bar", color="skyblue", figsize=(8,5))
plt.title("Average revenue by Region")
plt.xlabel("Region")
plt.ylabel("Average Revenue ($)")
plt.show()

plt.figure(figsize=(8,5))
df["Revenue"].hist(bins=10, edgecolor="black")
plt.title("Distribution of Revenue")
plt.xlabel("Revenue ($)")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x="Quantity", y="Revenue", data=df, hue="Region", s=100)
plt.title("Quantity vs Revenue")
plt.xlabel("Quantity Sold")
plt.ylabel("Revenue ($)")
plt.legend(title="Region")
plt.show()

print("Findings:")
print("- The dataset contains 10 rows and 5 columns.")
print("- No missing values were found.")
print("- The highest sales revenue comes from the west region.")
print("- Shoes generate higher revenue compared to Bags.")
print("- Revenue shows a positive trend with quantity (more items sold -> hogher revenue).")






    
    
```
